```python
#读取数据 
data=pd.read_excel('data.xlsx') 
data=data.iloc[:,:]
choose='BASPL'#选取这一列
data.tail()
```

# SVM+Elastic Net+Random Forest

```python
#支持向量机
from sklearn.svm import LinearSVR  
def svm_model():
    svr = LinearSVR() 
    model = svr.fit(X_train, y_train)
    #在训练集上的拟合结果
    y_train_predict=model.predict(X_train)
    y_test_predict=model.predict(X_test)
    return [y_train_predict,y_test_predict]
```

```python
#弹性网回归
from sklearn.linear_model import ElasticNet
def elasticnet_model():
    elasticnet=ElasticNet(alpha=.1, l1_ratio=.9, random_state=3)
    model = elasticnet.fit(X_train, y_train)
    #在训练集上的拟合结果
    y_train_predict=model.predict(X_train)
    y_test_predict=model.predict(X_test)
    return [y_train_predict,y_test_predict]
```

# SARIMA

```python
#sarima
#展示一下看看
data=pd.read_excel('data.xlsx') 
data=data[choose]
data=data/10000000#进行伪归一化
#构建训练集测试集
cut=6#取最后cut天为测试集
data=pd.Series(data.values,\
                             index=pd.DatetimeIndex(start='2015-01-01',end='2019-12-01',freq='MS'))
y=data[:-cut]
true=data[-cut:]
```

```python
#SARIMA模型 
def sarima_model():
    mod = sm.tsa.statespace.SARIMAX(y,
                                    order=(1, 1, 1),
                                    seasonal_order=(1, 1, 1, 12),
                                    enforce_stationarity=False,
                                    enforce_invertibility=False)
    results = mod.fit()
    #在训练集上进行预测
    pred = results.get_prediction(start=pd.to_datetime('2015-01-01'), dynamic=False)
    y_train_predict = pred.predicted_mean.values
    y_train = y.values
    y_train_predict = y_train_predict[1:]
    y_train_predict = y_train_predict[-42:]
    y_train = y_train[1:]
    #在测试集上进行预测
    pred_uc = results.get_forecast(steps=cut)  # retun out-of-sample forecast 
    y_test_predict=pred_uc.predicted_mean
    y_test_predict=y_test_predict.values
    y_test=true.values
    return [y_train_predict,y_test_predict]
```

```python
#构建神经网络数据集
y_train_predict4,y_test_predict4=sarima_model()
x_train_coef=np.array([y_train_predict1,y_train_predict2,y_train_predict3,y_train_predict4])
x_train_coef=x_train_coef.T
y_train_coef=y_train
x_test_coef=np.array([y_test_predict1,y_test_predict2,y_test_predict3,y_test_predict4])
x_test_coef=x_test_coef.T
```

```python
from sklearn.neural_network import MLPRegressor
# alpha:L2的参数：MLP是可以支持正则化的，默认为L2，具体参数需要调整
# hidden_layer_sizes=(5, 2) hidden层2层,第一层5个神经元，第二层2个神经元)，2层隐藏层，也就有3层神经网络
#clf = MLPRegressor(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(5, 2), random_state=1)
#'identity'，无操作激活，对实现线性瓶颈很有用，返回f（x）= x
#'logistic'，logistic sigmoid函数，返回f（x）= 1 /（1 + exp（-x））。
#'tanh'，双曲tan函数，返回f（x）= tanh（x）。
#'relu'，整流后的线性单位函数，返回f（x）= max（0，x）
NN= MLPRegressor(
    hidden_layer_sizes=(12,12,1),  activation='relu', solver='adam', alpha=0.0001, batch_size='auto',
    learning_rate='constant', learning_rate_init=0.001, power_t=0.5, max_iter=5000, shuffle=True,
    random_state=1, tol=0.0001, verbose=False, warm_start=False, momentum=0.9, nesterovs_momentum=True,
    early_stopping=False,beta_1=0.9, beta_2=0.999, epsilon=1e-08)
model = NN.fit(x_train_coef, y_train_coef)
y_test_predict=model.predict(x_test_coef)
```
